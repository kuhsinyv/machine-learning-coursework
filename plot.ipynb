```python
import os

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
history = []
with open("train.log", "r") as f:
    for line in f.readlines():
        line = line.split(",")
        for i in range(len(line)):
            line[i] = float(line[i].split(":")[1].strip()) if i != 0 else int(
                line[i].split(":")[1].strip()) # type: ignore
        history.append(line)
history = np.array(history)
history
```

```python
df = pd.DataFrame({
    "Epoch": history[:, 0].astype(int),
    "Loss": history[:, 1],
    "Training Accuracy": history[:, 2],
    "Test Accuracy": history[:, 3]
})
df
```

```python
plt.figure(figsize=(10,5), dpi=300)
plt.plot(df["Epoch"], df["Loss"])
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.tight_layout()
plt.savefig("loss.png")
```

```python
plt.figure(figsize=(10,5), dpi=300)
plt.plot(df["Epoch"], df["Training Accuracy"], label="Training Accuracy")
plt.plot(df["Epoch"], df["Test Accuracy"], label="Test Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.tight_layout()
plt.savefig("accuracy.png")
```
